```python
def apply_activation(input_tensor, activation_function):
    if activation_function == 'relu':
        return torch.nn.ReLU(input_tensor)
    if activation_function == 'identity':
        return input_tensor
    if activation_function == 'sigmoid':
        return torch.sigmoid(input_tensor)
```
